```python
#BT1 : mô phỏng không gian hoạt động cánh tay robot 2 bật tự do
import math as m
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
l1=50
l2=40
a=np.arange(0,m.pi/2,0.1)
b=np.arange(0,m.pi/2,0.1)

for t1 in a:
    for t2 in b:
            Px=(l1)*m.cos(t1)+l2*m.cos(t1+t2)
            Py=(l1)*m.sin(t1)+l2*m.sin(t1+t2)
            plt.xlabel('x(cm)')
            plt.ylabel('y(cm)')
            plt.plot(Px,Py,'m.')
plt.tight_layout()
plt.show()
print(t1,t2)
```

```python

l1=50
l2=40
l3=30
a=np.arange(0,m.pi,0.1)

for t1 in a:
    for t2 in a:
        for t3 in a:
            Px=(l1)*m.cos(t1)+l2*m.cos(t1+t2)+l3*m.cos(t1+t2+t3)
            Py=(l1)*m.sin(t1)+l2*m.sin(t1+t2)+l3*m.sin(t1+t2+t3)
            plt.xlabel('x(cm)')
            plt.ylabel('y(cm)')
            plt.plot(Px,Py,'c.')
plt.tight_layout()
plt.show()
```

```python
#BT3: biểu đồ theta 1,2,3, khoảng cách trục l2 theo thời gian tt
l1=50
l3=20
v=10
p=10
h=100

for t in np.arange(0,10,0.1):
    t1 = m.atan2((h - p - l3), v * t)
    t2 = m.pi / 2 - t1
    l2 = m.sqrt(((h - p - l3) ** 2) + ((v * t) ** 2)) - l1
    Px = (l1 + l2) * m.cos(t1) + l3 * m.cos(t1 + t2)
    Py = (l1 + l2) * m.sin(t1) + l3 * m.sin(t1 + t2)
    plt.subplot(2,2,1)
    plt.plot(Px, Py, 'r.')
    plt.title('Đồ thị 1')
    plt.xlabel('Y(cm)')
    plt.ylabel('X(cm)')
    plt.subplot(2, 2, 2)
    plt.plot(t, t1, 'y.')
    plt.title('Đồ thị 2')
    plt.ylabel('theta1(rad)')
    plt.xlabel('time(s)')
    plt.subplot(2, 2, 3)
    plt.plot(t, t2, 'b.')
    plt.title('Đồ thị 3')
    plt.ylabel('theta2(rad)')
    plt.xlabel('time(s)')
    plt.subplot(2, 2, 4)
    plt.plot(t, l2,'g.' )
    plt.title('Đồ thị 4')
    plt.ylabel('L2')
    plt.xlabel('time(s)')
plt.tight_layout()
plt.show()
```

```python
#BT4: bài toán đọng học nghịch cánh tay robot 2 bậc tự do
xa = 60
ya = 30
xb = 20
yb = 60
l1 = 50
l2 = 40
plt.figure(figsize=(10,5))
for t in np.arange(0,2*np.pi,0.01):
    #hình trái tim
    x=m.sin(t)
    y=(m.cos(t)+abs(m.sin(t)**(2/3)))
    #đường tròn
    '''x = m.sin(t)
    y = m.cos(t)'''
    #tịnh tuyến
    '''x = xa + (xb - xa) * t
    y = ya + (yb - ya) * t'''
    c2 = (x ** 2 + y ** 2 - l1 ** 2 - l2 ** 2)/(2*l1*l2)
    s2 = m.sqrt(abs(1 - c2 ** 2))
    t2 = m.atan2(s2, c2)

    c1 = (x * (l1 + l2 * c2) + y * (l2 * s2))
    s1 = (y * (l1 + l2 * c2) - x * (l2 * s2))
    t1 = m.atan2(s1, c1)
    Px = l1 * m.cos(t1) + l2 * m.cos(t1 + t2)
    Py = l1 * m.sin(t1) + l2 * m.sin(t1 + t2)
    plt.subplot(1, 2, 1)
    plt.plot(t, t1*180/m.pi , 'r.')
    plt.plot(t, t2*180/m.pi , 'c.')
    plt.legend(['t1','t2'])
    plt.subplot(1, 2, 2)
    plt.plot(Px, Py, 'm.')
plt.show()



```

```python
#BT5: bài tập về nhà tuần 5 ( di chuyển vị trí A(30,30,30)đến B(70,70,30))
l1=40
l2=70
l3=60
ax = plt.axes(projection="3d")
fig = plt.figure(figsize=(10, 10))
a=np.arange(0,1,0.01)
for t in a:
    x = 30+40*t
    y = 30+40*t
    z = 30
    t1=m.atan2(-x,y)
    A=-x/m.sin(t1)
    B=z-l1
    c3=(A**2+B**2-l2**2-l3**2)/(2*l2*l3)
    s3=m.sqrt(m.fabs(1-c3**2))
    t3=m.atan2(s3,c3)
    #t3 = m.acos((A**2+B**2-l2**2-l3**2)/(2*l2*l3))
    c2=(A*(l2+l3*c3)+B*l3*s3)/((l2+l3*c3)**2-(l3*s3)**2)
    s2=(B*(l2+l3*c3)-A*l3*m.sin(t3))/((l2+l3*c3)**2-(l3*s3)**2)
    t2=m.atan(s2/c2)

    Px=-m.sin(t1)*(l2*m.cos(t2)+l3*m.cos(t2+t3))
    Py=m.cos(t1)*(l2*m.cos(t2)+l3*m.cos(t2+t3))
    Pz=l1+l2*m.sin(t2)+l3*m.sin(t2+t3)
    ax.scatter3D(Px, Py, Pz )

    plt.subplot(2, 2, 1)
    plt.plot(t, t1, 'y.')
    plt.title('theta1')
    plt.subplot(2, 2, 2)
    plt.plot(t, t2, 'm.')
    plt.title('theta2')
    plt.subplot(2, 2, 3)
    plt.plot(t, t3, 'c.')
    plt.title('theta3')
plt.show()
```

```python
#BT6: mô phỏng cánh tay robot sơn bên trong thùng rỗng
R = 90
p=10
ax = plt.axes(projection="3d")
a=np.arange(0,40*m.pi,0.1)
for t in a:
    x = R*m.cos(t)
    y = R*m.sin(t)
    z=p/2*m.pi*t
    t1=m.atan2(-x,y)
    l1=z
    l2=m.sqrt(x**2+y**2)
    Px=-l2*m.sin(t1)
    Py=l2*m.cos(t1)
    Pz=l1
    ax.scatter3D(Px, Py, Pz )
plt.show()
```
